```python
%load_ext autoreload
%autoreload 2
%reload_ext autoreload
import sys
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import polars as pl
sys.path.append("/Users/gabriel/Desktop/Quant_Project/Quant-Summer-Project")

from data.fetcher import fetch_data
from data.cleaner import clean_data
from data.loader import add_event_column,filter_summer
from config.settings import HISTORICAL_START,HISTORICAL_END,DEFAULT_CITY

df_raw = fetch_data(HISTORICAL_START,HISTORICAL_END)


df_clean = clean_data(df_raw)


df_event = add_event_column(df_clean)






```

```python
df_temp = df_event["temperature_2m_max"].to_list()
df_year = df_event["time"].to_list()

plt.figure(figsize =(30,10))
plt.title(f"temperature_2m_max from {HISTORICAL_START} to {HISTORICAL_END} in {DEFAULT_CITY}")
plt.xlabel("date")
plt.ylabel("temperature_2m_max")


sns.lineplot(x = df_year , y = df_temp)
plt.xticks(ticks=range(0, len(df_year), 365), labels=[df_year[i][:4] for i in range(0, len(df_year), 365)], rotation=45)
```

We can clearly see similar temperature cycles every year. Some years show a slight difference in minimum temperature, but they also tend to peak slightly higher in maximum temperature compared to the average. In general, the temperature averages around 10°C, so this is the graph's baseline. This pattern repeats every year, reflecting the different seasons. Consequently, this allows us to predict future temperatures to some extent

```python
plt.figure(figsize =(40,15))
plt.title(f"temperature_2m_max from {HISTORICAL_START} to {HISTORICAL_END} in {DEFAULT_CITY} Histogram")
plt.xlabel("temperature_2m_max")
sns.histplot(  x= df_temp,binwidth=0.5,kde =True)
```

The data shows two peaks, which tells us this is a bimodal distribution. It rises from -10°C to approximately 6°C, drops back down, and then starts to rise again at 14°C, creating the second peak. It then reaches 20°C before dropping down again. This implies that certain models will perform poorly, specifically the Gaussian model. Consequently, we will build multiple models, using the assumption that a single Gaussian model will perform badly as our baseline.
The reason this bimodal distribution appears is that we are mixing all the seasons together. Winter and summer create the two distinct peaks, while autumn and spring account for the temperatures in between and this is why the distribution dips between the two peaks.

```python

df_temp_summer = filter_summer(df_event)
df_temp_list = df_temp_summer["temperature_2m_max"].to_list()

plt.figure(figsize =(40,15))
plt.title(f"temperature_2m_max from {HISTORICAL_START} to {HISTORICAL_END} in {DEFAULT_CITY} Histogram (June - August) ")
plt.xlabel("temperature_2m_max")
sns.histplot(  x= df_temp_list,binwidth=0.5,kde =True)
                                                       


```

When looking only at the summer months compared to a full year, we get a bell curve (normal distribution) instead of a bimodal distribution. This occurs because the summer months represent only a single season, which does not contain seasonal cycles. Consequently, the Gaussian baseline model works better when we isolate the summer months rather than accounting for the entire year. However, we can see that it is not a perfect bell curve, as it is skewed when we have hotter days.
